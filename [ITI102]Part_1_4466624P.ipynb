**Question 1. How many men and women (*sex* feature) are represented in this dataset? (1 mark)**


````python
#answer

# select 'sex' column, drop na and make it case in-sensitive
gender_counts = data["sex"].dropna().str.lower().value_counts()

# get counts
female = gender_counts["female"]
male = gender_counts["male"]

print("Number of Female:", female)
print("Number of Male:", male)
````

**Question 2. What is the average age (*age* feature) of women? (1 mark)**

````python
#Answer

# get all 'age' where sex = 'Female' (case in-sensitive), and strip it
woman_and_age_data = data[data["sex"].str.lower().str.strip() == "female"]["age"]

# calculate average of woman age
average_women_age = woman_and_age_data.mean()
print(f"Average age of women: {average_women_age:.2f}")
````

**Question 3. What is the proportion of German citizens (*native-country* feature)? ( 1 mark)**

````python
#Answer

# get all records with native-country = german
german_data = data[data["native-country"].str.lower() == "germany"]["native-country"]

# calculate proportion of german citizens
total_records = len(data)
print(f"Total number of records: {total_records}")
total_german  = len(german_data)
print(f"Total number of records with native-countr=German: {total_german}")

proportion_german = (total_german / total_records) * 100
print(f"Proportion of German citizens is {total_german} out of {total_records}, about {proportion_german:.2f}% of total records")
````


**Question 4.What are mean value and standard deviation of the age for those who recieve more than 50K per year (*salary* feature) and those who receive less than 50K per year? ( 2 marks)**

````python
#anwser

# get all ages and salary, upper case it and strip it
age_and_salary_data = data[["age", "salary"]]
age_and_salary_data.loc[:, 'salary'] = age_and_salary_data["salary"].str.upper().str.strip()

# calculate mean and std for age based on salary
age_stats = age_and_salary_data.groupby("salary")["age"].agg(["mean", "std"])
age_stats = age_stats.round(2) # rounded to 2 decimals
age_stats.columns = ["mean_age", "std_age"]
age_stats
````

**Question 5. Compare the education of people who receive more than 50k  with poeple who receive less then and equal 50k. Show and comment your comparsion. (2 marks)**

````python
#answer

# get all education and salary, make it case in-sensitive, and strip it
education_income_data = data[["education", "salary"]]
education_income_data.loc[:, "education"] = education_income_data["education"].str.upper().str.strip()
education_income_data.loc[:, "salary"] = education_income_data["salary"].str.upper().str.strip()


# Group by education and salary, count records
grouped = education_income_data.groupby(["education", "salary"]).size().unstack(fill_value=0)

# Calculate percentage share per education level
grouped["% >50K"] = (grouped[">50K"] / grouped.sum(axis=1)) * 100
grouped["% <=50K"] = (grouped["<=50K"] / grouped.sum(axis=1)) * 100
grouped = grouped.round(2)


# Sort by percentage of high-income individuals
education_income_sorted = grouped.sort_values("% >50K", ascending=False)

# Show the results
print(education_income_sorted[[">50K", "<=50K", "% >50K", "% <=50K"]])

print("The data shows a clear positive correlation between education level and income.")
print("It shows the individuals with higher education level like Doctorates, Professional, and Master's are far more likely to earn over $50K")
print("THose with lower education levels—especially below Bachelor Degree earn less than $50K.")
````

**Question 7. What is the maximum number of hours a person works per week (*hours-per-week* feature)? What is the percentage of people work maximum number of hour per week and who earn >50K? (2 marks)**

````python
#answer

work_data = data[["hours-per-week", "salary"]]
total_workers = len(work_data)

max_hours = work_data["hours-per-week"].max()
total_workers_max_hours = len(work_data[work_data["hours-per-week"] == max_hours])
percentage_total_workers_max_hours = (total_workers_max_hours / total_workers) * 100
print(f"Maximum number of hours a person works per week: {max_hours}")
print(f"Total number of people who work maximum number of hours per week: {total_workers_max_hours}")
print(f"Percentage of people who work maximum number of hours per week: {percentage_total_workers_max_hours:.2f}%")
````

**Question 9. Obtain the Skewness from the Question 8 histogram plot. Comment on the skewness. Suggest how to improve on it. (2 marks)**
